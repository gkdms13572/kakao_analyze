# 보낸 사람 기준 빈도수 세기 1


## 1. txt 파일 불러오기 (열기)

```python
Kakao = open("/content/대화내용 (1).txt", "r", encoding="UTF-8")
```

## 2. 파일 한줄씩 읽어서 list로 변환 -> 카톡내용을 분석하기 위해 python에서 불러옴!

```python
 li_Kakao = []

for line in Kakao:
    if line != "\n": # 불필요한 줄바꿈 제거
        li_Kakao.append(line)

print(li_Kakao)
```

## 3. 톡 하나에서 이름만 추출하기

```python
li_send = [] # 보낸사람 이름만 저장
for talk in li_Kakao:
    start = talk.find("[") +1
    end = talk.find("]")
    send = talk[start : end]
    # 어떤 데이터가 불필요한 데이터일까???
    # 글자수가 너무 적거나 너무 많으면 불필요함!
    # 글자수가 2개 이하이거나 6개 이상이면 출력하지 않겠다!
    if len(send) > 2 and len(send) <= 6 : # 글자수가 3보다 크고 5보다 작다면!
        # print("보낸사람!! -> ", send) # 출력하겠다!
        li_send.append(send) # 리스트에 추가~

print(li_send)
```

## 4. 빈도수 세기

```python
# - 중복을 제거한 범주를 알아낸다. -> 리스트에서 중복제거 하는법 ! (set)
# - 각 범주별로 값이 몇개씩 있는지 확인! -> 리스트에서 값이 몇개 들어있는지 알아내는 법! (count)
이름들 = set(li_send)
print(이름들)

for data in 이름들: # 이름 범주에서 하나씩 꺼내서
    cnt = li_send.count(data) # 모든 단톡방 보낸사람 데이터 중 몇개 있는지 확인
    print(data, cnt) # 이름1개, 이름이 나온 횟수
```

# 보낸 사람 기준 빈도수 세기 2


## 1. 파일 불러오기

```python
#1. 파일 불러오기

from google.colab import files

# 파일 업로드
uploaded = files.upload()

# 업로드된 파일 이름 확인
for filename in uploaded.keys():
    print(f'Uploaded file name: {filename}')
```

Google Colab에서 files.upload()를 호출하면, 업로드된 파일들은 딕셔너리 형태로 저장됩니다. 이 딕셔너리는 파일 이름을 키로 하고, 해당 파일의 내용을 바이트 배열로 저장합니다.

  ## 2. 파일 내용 읽기

```python
# 업로드된 파일 내용을 읽기 위한 작업
file_content = uploaded[filename].decode('utf-8')  # 파일 내용을 문자열로 디코딩

# 파일 내용을 한 줄씩 읽어서 리스트로 변환
li_kakao = file_content.splitlines()  # 문자열을 줄 단위로 분리하여 리스트로 변환

# 새로운 리스트 생성
filtered_lines = [line for line in li_kakao if line.strip()]


# 결과 출력
print(filtered_lines)
```

## 3. for문과 in연산자, if문을 통해 전처리

```python
kakao_new = []
str1 = "서하은"
str2 = "달🏖숭"
for kakao in li_kakao_new:
  if str1 in kakao:
    kakao_new.append(kakao)
  elif str2 in kakao:
    kakao_new.append(kakao)
print(kakao_new)
```

## 4. count함수를 사용하여 빈도수 세기

```python
seo_cnt = int(kakao_new.count("서하은"))
print(seo_cnt)
kakao_new.count("달🏖숭")
park_cnt = int(kakao_new.count("달🏖숭"))
print(park_cnt)
```

# 대화내용 전부다 엑셀파일로 저장하기

```python
# 대화내용 전부다 엑셀파일로 저장하기
file = open("/content/대화내용 (1).txt", "r", encoding = "UTF-8")

li_kakao = []
for line in file:
   if line != "":
       li_kakao.append(line)


li_send = []

# 전처리
for data in li_kakao:
    name = data[data.find("["):data.find("]")+1]
    if name != "":
        data = data.replace(name, "")
        date = data[data.find("["):data.find("]")+1]
        msg = data[data.find("]")+1:]
        dic = {"name" : name, "date" : date, "msg" : msg}
        li_send.append(dic)

print(li_send)


import openpyxl as op

workbook = op.Workbook()

sheet = workbook.active

sheet["A1"].value = "보낸사람"
sheet["B1"].value = "시간"
sheet["C1"].value = "내용"

i = 3
for data in li_send:
    sheet.cell(row = i, column = 1).value = data.get("name")
    sheet.cell(row = i, column = 2).value = data.get("date")
    sheet.cell(row = i, column = 3).value = data.get("msg")
    i = i+1

workbook.save("/content/카톡대화.xlsx")
```
